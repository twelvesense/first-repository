```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '/content/drive/MyDrive/data/dataton/population.csv'
data = pd.read_csv(file)
data
```

```python
df = data.set_index(keys=['population'])
df
```

```python
total = df.loc['total']
male = df.loc['male']
female = df.loc['female']
```

```python
year = np.arange(1925, 2021, 5)
year
```

```python
log_total = np.log(total)
log_male = np.log(male)
log_female = np.log(female)
```

```python
plt.figure(figsize=(20,12))

plt.plot(year, log_total, label='total', color='black')
plt.plot(year, log_male, label='male', linestyle='dotted', color='navy')
plt.plot(year, log_female, label='female', linestyle='dashed', color='violet')

plt.legend()
plt.show()
```
